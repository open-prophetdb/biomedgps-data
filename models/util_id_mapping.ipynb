```python
import pandas as pd

data = pd.read_csv('./examples/mecfs.tsv', sep='\t')
```

```python
entities = pd.read_csv('../graph_data/entities.tsv', sep='\t')
```

```python
# Return matched line if a id matches with the id or xrefs in the entities.tsv
def get_matched_id(id, etype, entities):
    rows = entities[((entities["id"] == id) | entities["xrefs"].str.contains(id)) & (entities["label"] == etype)]
    if len(rows) == 0:
        return None
    else:
        return rows.iloc[0]["id"]

def get_matched_name(id, etype, entities):
    rows = entities[((entities["id"] == id) | entities["xrefs"].str.contains(id)) & (entities["label"] == etype)]
    if len(rows) == 0:
        return None
    else:
        return rows.iloc[0]["name"]
```

```python
source_ids = data["source_id"].to_list()
source_types = data["source_type"].to_list()
data['matched_source_id'] = [get_matched_id(id, etype, entities) for id, etype in zip(source_ids, source_types)]
data['matched_source_name'] = [get_matched_name(id, etype, entities) for id, etype in zip(source_ids, source_types)]

target_ids = data["target_id"].to_list()
target_types = data["target_type"].to_list()
data['matched_target_id'] = [get_matched_id(id, etype, entities) for id, etype in zip(target_ids, target_types)]
data['matched_target_name'] = [get_matched_name(id, etype, entities) for id, etype in zip(target_ids, target_types)]
```

```python
data.to_csv('./examples/mecfs_matched.tsv', sep='\t', index=False)
```
